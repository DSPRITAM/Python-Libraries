# Seaborn

```python
# Seaborn is an amazing visualization library for statistical graphics plotting in Python.
# It provides beautiful default styles and color palettes to make statistical plots more attractive.
```

```python
#!pip install seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## set(style = )

```python
# Histplot

x = np.random.normal(10,20,50)

sns.set(style='darkgrid')
# style could be white,dark,whitegrid,darkgrid,ticks.

sns.histplot(x,kde=True)
plt.show()
```

```python
x = np.array([10,12,15,12,18])
y = np.array([45,43,47,44,49])

sns.set(style='dark')
sns.histplot(x=x,y=y)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.info()
```

```python
sns.set(style='white')
sns.histplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# Countplot
# we generally use countplot to plot categorical data on x axis.

sns.set(style = 'whitegrid')

sns.countplot(x ='sex', data = tips)
sns.despine()# it will remove the upper and right side border of the chart
```

```python
sns.countplot(x='smoker',data=tips)
plt.show()
```

```python
plt.figure(figsize = (8,4))  #8 is width and 4 is height.
#with the help of figure(figsize) we can adjust the figure size.

sns.set(style = 'dark')
sns.countplot(x = 'day', data=tips)
plt.show()
```

```python
sns.countplot(x = 'time', data=tips)
plt.show()
```

```python
# Linear model plot

sns.lmplot(x='total_bill',y='tip',height=4,aspect=2,data=tips)
# where,
# x = total_bill,
# y = tip,
# height = 4 we can specify as per requirement,
# aspect = 2 width = 2 times of its height,
# data = tips
```

```python
tips.info()
```

```python
sns.lmplot(x='total_bill',y='size',data=tips,height=3,aspect=2)
plt.show()
```

```python
sns.lmplot(x='size',y='tip',data=tips,height=3,aspect=2)
```

```python
# Scatterplot
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips)
plt.show()
```

```python

sns.scatterplot(x='total_bill', y='tip', hue='smoker',data=tips)
```

```python
colors={'Yes':'Green',
       'No':'Red'}
sns.scatterplot(x='total_bill',y='tip',hue='smoker',palette=colors, data=tips)
plt.show()
```
