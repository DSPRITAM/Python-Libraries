## Markers

```python
# we can use marker keyword to emphasize each point with special style.

# by default x axis will start from [0,1,2,3,4...] if we didnt specify
y = np.array([3,6,5,8])

plt.plot(y,marker = 'o')
plt.show()
```

```python
y = np.array([3,6,5,8])

plt.plot(y,marker = 'p')
plt.show()
```

```python

# lineMarker Reference
# You can choose any of these markers:

# Marker	Description
# 'o'	Circle	
# '*'	Star	
# '.'	Point	
# ','	Pixel	
# 'x'	X	
# 'X'	X (filled)	
# '+'	Plus	
# 'P'	Plus (filled)	
# 's'	Square	
# 'D'	Diamond	
# 'd'	Diamond (thin)	
# 'p'	Pentagon	
# 'H'	Hexagon	
# 'h'	Hexagon	
# 'v'	Triangle Down	
# '^'	Triangle Up	
# '<'	Triangle Left	
# '>'	Triangle Right	
# '1'	Tri Down	
# '2'	Tri Up	
# '3'	Tri Left	
# '4'	Tri Right	
# '|'	Vline	
# '_'	Hline
```

```python
# marker|line|color
```

```python
y = np.array([1,4,7,3])

plt.plot(y,'o--b')
plt.show()

# where,
# o for points,-- for double dash line,b for blue color
```

```python
# Marker size

y = np.array([2,4,3,7])
plt.plot(y,marker = "o" , ms = 20) # ms=marker size = 20.
plt.show()
```

```python
# we can even specify the marker color
y = np.array([4,3,7,8])
plt.plot(y, marker = "o", ms=30, mec='r') # mec color the edges only.
plt.show()
```

```python
y = np.array([4,3,7,8])
plt.plot(y, marker = "o", ms=30, mfc='r') # mfc color the inside only.
plt.show()
```

```python
# if we want to change the whole color of point
y = np.array([4,3,7,8])
plt.plot(y, marker = "o", ms=30, mec='r', mfc='r')
plt.show()
```

## Linestyle

```python
# we can change the linestyle by using linestyle or ls keyword
```

```python
y = np.array([2,7,4,6])

plt.plot(y,marker = 'o',linestyle = "dotted")
plt.show()
```

```python
y = np.array([2,6,4,3])
plt.plot(y , marker = 'o', ls=':')
plt.show()
```

```python
# You can choose any of these styles:

#      Style	          Or
# 'solid' (default)   	'-'	
# 'dotted'            	':'	
# 'dashed'            	'--'	
# 'dashdot'           	'-.'	
# 'None'              	'' or ' '
```

```python
# Linewidth


y = np.array([4,6,5,9])
plt.plot(y,ls='-',linewidth=20) # we can specify linewidth according to our requirement.
plt.show()
```

```python
# Multiple line

x = np.array([2,4,6,8])
y1 = np.array([2,6,4,8])
y2 = np.array([11,15,13,18])

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x = np.array([2,4,6,8])
y1 = np.array([2,7,4,9])
y2 = np.array([13,6,12,18])

plt.plot(x,y1,x,y2)
plt.show()
```

 ## Labels

```python
x = np.array([10,11,12,13])
y = np.array([21,22,23,24])

plt.plot(x,y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

## Subplot

```python
# display multiple plots
```

```python
x = np.array([0,1,2,3,4])
y = np.array([10,11,12,13,14])

plt.subplot(1,2,1)
# The figure has 1 row, 2 columns and this is first plot
plt.plot(x,y)


x = np.array([0,1,2,3,4])
y = np.array([10,20,30,40,50])
plt.subplot(1,2,2)
# The figure has 1 row, 2 columns and this is second plot
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0,1,2,3,4])
y = np.array([10,11,12,13,14])

plt.subplot(2,2,1)
# The figure has 2 row, 2 columns and this is first plot
plt.plot(x,y)


x = np.array([0,1,2,3,4])
y = np.array([10,20,30,40,50])
plt.subplot(2,2,2)
# The figure has 2 row, 2 columns and this is second plot
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0,1,2,3,4])
y = np.array([10,11,18,13,14])

plt.subplot(3,2,1)
# The figure has 3 row, 2 columns and this is first plot
plt.plot(x,y)
plt.title("SALES")


x = np.array([0,1,2,3,4])
y = np.array([100,20,30,40,50])
plt.subplot(3,2,2)
# The figure has 3 row, 2 columns and this is second plot
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

## SCATTER PLot

```python
x = np.array([1,2,3,4,9,5,3,7,23,6])
y = np.array([100,120,145,124,154,165,154,18,154,135])

plt.scatter(x,y)
plt.show()
```

```python
x = np.array([10,11,12,10,12,10,13,12,11,12])
y = np.array([25,24,30,22,24,23,32,24,21,23])
# Age and weight of boys in class 4th...blue color
plt.scatter(x,y)

x = np.array([11,12,11,12,11,10,11,12,12,12])
y = np.array([20,18,17,18,19,20,20,18,16,18])
# Age and weight of girls in class 4th...orange color
plt.scatter(x,y)
plt.show()
```

```python
# Colors

# we can set colors for the scatter plot

x = np.array([10,11,12,10,12,10,13,12,11,12])
y = np.array([25,24,30,22,24,23,32,24,21,23])
# Age and weight of boys in class 4th...green color
plt.scatter(x,y, color = 'green')

x = np.array([11,12,11,12,11,10,11,12,12,12])
y = np.array([20,18,17,18,19,20,20,18,16,18])
# Age and weight of girls in class 4th...red color
plt.scatter(x,y, color = 'red')
plt.show()
```

```python
# color every dot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, color = colors)

plt.show()
```

## Histogram

```python
# A histogram is a graph showing frequency distributions.

# It is a graph showing the number of observations within each given interval.

x = np.random.randint(100,200,50)
# start point is 100 , end point is 200 and 50 is the frequency

plt.hist(x)
plt.show()
```

```python
x = np.random.normal(150, 10, 500)
# normal data distribution by numpy

plt.hist(x)
plt.show() 
```

## Pie chart

```python
x = np.array([10,20,14,22])

plt.pie(x)
plt.show()
```

```python
# Note: The size of each wedge is determined by comparing the value with all the other values, by using this formula:

# The value divided by the sum of all values: x/sum(x)
```

```python
y = np.array([200,40,500,300])
Fruits = ["Apple","Banana","Kiwi","Guava"]

plt.pie(y, labels = Fruits, startangle=90)
# As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

# The startangle parameter is defined with an angle in degrees, default angle is 0:

plt.show()
```

```python
# Explode

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode , shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels)
plt.legend(title = 'four fruits',loc=10)
# location can be 'best'    0
#         'upper right'     1
#         'upper left'      2
#         'lower left'      3
#         'lower right'     4
#         'right'           5
#         'center left'     6
#         'center right'    7
#         'lower center'    8
#         'upper center'    9
#         'center'          10
plt.show() 
```
